```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#generate range of dates for polling data frame
dates = pd.date_range('20160601', periods = 160, freq = 'D')

#simulate some noisy polling data
Poll_data = np.random.normal(loc = 0.0, scale = 1.0, size = len(dates))*0.035 + 0.06

#create data frame
df = pd.DataFrame(Poll_data, index = dates)
```

```python
df.plot()
plt.show()
```
